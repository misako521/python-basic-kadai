```python
# timeモジュールをインポート
import time

# 現在の日時を指定したフォーマットで出力する
print(time.strftime("%Y年%m月%d日%H時%M分%S秒", time.localtime()))
```

```python
class Human:
  def __init__(self, name, age):
      self.name = name
      self.age = age

  def check_adult(self):
      if self.age >= 20:
          print(f"{self.name}は大人")
      else:
          print(f"{self.name}は大人ではない")

user1 = Human("misako", 2025-1992)
user2 = Human("misa", 20)
user3 = Human("mimi", 10*5)
user5 = Human("nano", -1)

users = [user1, user2, user3,user5]

for user in users:
    user.check_adult()
```
