```python
# timeモジュールをインポート
import time

# 現在の日時を指定したフォーマットで出力する
print(time.strftime("%Y年%m月%d日%H時%M分%S秒", time.localtime()))
```

```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age


    def printinfo(self):
        print(f"名前：{self.name}")
        print(f"年齢：{self.age}")

user = Human("misako", 33)

user.printinfo()
```
